```python
def rastrigin(X, *args):
    return sum(list(map(lambda x: x**2 - 10*np.cos(2*np.pi*x) - 10, list(X))))
```

```python
X = [np.arange(-12,12,0.2), np.arange(-12,12,0.2)]
```

```python
ACOsearch = ACO(1, 1, 0.5, 1)
ACOsearch.optimize(5, 5, X, rastrigin, verbose=True)
```
